```python
# -*- coding: utf-8 -*-
"""
Create all processed datasets from raw inputs.

Inputs (expected in data/raw/):
- Unemployment rate.csv
- Key statistics Seasonally adjusted.csv
- Data1-Table 1.csv                 # (optional fallback for CPI headline)
- Table 3.1-Table 1.csv             # CPI components (e.g., Groups)
- Table 3.3-Table 1.csv             # CPI components (e.g., Sub-groups)
- Table 3.5-Table 1.csv             # CPI components (Selected Expenditure classes)
- quarterly_personal_insolvencies.csv
- regional_quarterly_time_series.csv

Outputs (written to data/processed/):
- unemployment_quarterly.csv
- cpi_quarterly.csv
- cpi_components_yoy.csv
- afsa_insolvencies_quarterly.csv
- afsa_insolvencies_state_quarterly.csv
- merged_macro.csv
- final_dataset.csv
"""
import os
import re
import numpy as np
import pandas as pd

ROOT = os.path.dirname(os.path.dirname(__file__))            # project-root/src -> project-root
RAW  = os.path.join(ROOT, "data", "raw")
PROC = os.path.join(ROOT, "data", "processed")
os.makedirs(PROC, exist_ok=True)

# --- Helpers -----------------------------------------------------------------

def to_quarter_period(s, default_freq="Q-DEC"):
    """
    Normalize many quarter formats to pandas Period with Australian quarters (Mar/Jun/Sep/Dec).
    Accepts '2021-Q3', 'Q3 2021', 'Sep-2021', '2021 Sep', '2021-Q3 (seasonally adjusted)', etc.
    """
    if pd.isna(s):
        return pd.NaT
    s = str(s).strip()

    # explicit 2021-Q3 like
    m = re.search(r"(\d{4})\s*[- ]?Q([1-4])", s, flags=re.I)
    if m:
        y, q = int(m.group(1)), int(m.group(2))
        return pd.Period(f"{y}Q{q}", freq=default_freq)

    # month-based (ABS quarters end Mar/Jun/Sep/Dec)
    m2 = re.search(r"(\d{4}).*?(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)", s, flags=re.I)
    if m2:
        y = int(m2.group(1))
        mon = m2.group(2).title()
        month_num = {"Jan":1,"Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,
                     "Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}[mon[:3]]
        dt = pd.Timestamp(year=y, month=month_num, day=1)
        return dt.to_period(default_freq)

    # bare year -> NaT (we need quarter granularity)
    return pd.NaT


def detect_col(df, patterns, default=None):
    """Pick the first column whose name matches any of the regex patterns."""
    cols = list(df.columns)
    for pat in patterns:
        for c in cols:
            if re.search(pat, str(c), flags=re.I):
                return c
    return default


def clean_numeric(series):
    """Turn strings with %/commas/spaces into float."""
    return (
        series.astype(str)
              .str.replace(r"[,%\s]", "", regex=True)
              .replace({"": np.nan, "nan": np.nan})
              .astype(float)
    )


def yoy(series):
    """Year-on-year percentage change for a quarterly series."""
    return series.pct_change(4) * 100.0


# --- 1) Unemployment (national) ----------------------------------------------

def build_unemployment_quarterly():
    path = os.path.join(RAW, "Unemployment rate.csv")
    df = pd.read_csv(path)

    # Try to find time & value columns
    time_col = detect_col(df, [r"^time$", r"date", r"period", r"quarter", r"ref.*period"])
    val_col  = detect_col(df, [r"unemploy", r"rate", r"value", r"^sa$", r"seasonally.*adj"])

    if time_col is None or val_col is None:
        # Attempt wide-to-long if needed
        # e.g., first col is period, others are series
        if df.shape[1] > 2:
            df = df.melt(id_vars=[df.columns[0]], var_name="series", value_name="value")
            time_col = df.columns[0]
            val_col  = "value"
        else:
            raise ValueError("Could not detect time/value columns in Unemployment rate.csv")

    out = pd.DataFrame({
        "quarter": df[time_col].map(lambda x: to_quarter_period(x)),
        "unemployment_sa": clean_numeric(df[val_col]),
    }).dropna(subset=["quarter"]).sort_values("quarter")

    out["unemployment_sa_yoy_pp"] = yoy(out["unemployment_sa"])  # YoY change in percentage points (approx.)
    out.to_csv(os.path.join(PROC, "unemployment_quarterly.csv"), index=False)
    return out


# --- 2) CPI headline (national) ----------------------------------------------

def build_cpi_quarterly():
    # Prefer Key statistics Seasonally adjusted.csv; fall back to Data1-Table 1.csv
    candidates = [
        os.path.join(RAW, "Key statistics Seasonally adjusted.csv"),
        os.path.join(RAW, "Data1-Table 1.csv"),
    ]
    for path in candidates:
        if os.path.exists(path):
            cpi = pd.read_csv(path)
            break
    else:
        raise FileNotFoundError("No CPI headline file found. Expected one of the two candidates.")

    time_col = detect_col(cpi, [r"^time$", r"date", r"period", r"quarter"])
    val_col  = detect_col(cpi, [r"cpi", r"all groups", r"index", r"value", r"^sa$"])

    if time_col is None or val_col is None:
        # try wide → long
        if cpi.shape[1] > 2:
            cpi = cpi.melt(id_vars=[cpi.columns[0]], var_name="series", value_name="value")
            time_col = cpi.columns[0]; val_col = "value"
        else:
            raise ValueError("Could not detect time/value columns in CPI file")

    out = pd.DataFrame({
        "quarter": cpi[time_col].map(lambda x: to_quarter_period(x)),
        "cpi_index_sa": clean_numeric(cpi[val_col]),
    }).dropna(subset=["quarter"]).sort_values("quarter")

    out["cpi_yoy_pct"] = yoy(out["cpi_index_sa"])
    out.to_csv(os.path.join(PROC, "cpi_quarterly.csv"), index=False)
    return out


# --- 3) CPI components (YoY) -------------------------------------------------

def build_cpi_components_yoy():
    component_files = [
        ("Table 3.1-Table 1.csv", "Group"),
        ("Table 3.3-Table 1.csv", "Subgroup"),
        ("Table 3.5-Table 1.csv", "Class"),
    ]
    frames = []
    for fname, level in component_files:
        path = os.path.join(RAW, fname)
        if not os.path.exists(path):
            continue
        df = pd.read_csv(path)

        # Heuristics: first column ~ component name, one column ~ time or data in wide format
        # Handle both long and wide.
        # Try to detect if rows are time and columns are components (wide):
        if df.columns.str.contains(r"\b(20\d{2}|19\d{2})\b", case=False, regex=True).any():
            # Likely wide with time-like headers — transpose logic
            df_long = df.melt(id_vars=[df.columns[0]], var_name="period", value_name="value")
            comp_col = df.columns[0]
            period_col = "period"
            val_col = "value"
        else:
            # Otherwise assume already long-ish
            comp_col = detect_col(df, [r"(group|sub.*group|class|component|series)"], default=df.columns[0])
            period_col = detect_col(df, [r"^time$", r"date", r"period", r"quarter"])
            val_col = detect_col(df, [r"value|index|price|cpi"], default=df.columns[-1])

            if period_col is None:
                # Try melt with the first col = component; rest are periods
                df_long = df.melt(id_vars=[comp_col], var_name="period", value_name="value")
                period_col = "period"; val_col = "value"
            else:
                df_long = df[[comp_col, period_col, val_col]].copy()

        df_long.rename(columns={comp_col: "component", period_col: "period", val_col: "value"}, inplace=True)
        df_long["quarter"] = df_long["period"].map(to_quarter_period)
        df_long["value"]   = clean_numeric(df_long["value"])
        df_long = df_long.dropna(subset=["quarter", "component"]).sort_values(["component", "quarter"])

        # YoY per component
        df_long["yoy_pct"] = df_long.groupby("component", group_keys=False)["value"].apply(yoy)

        df_long["level"] = level
        frames.append(df_long[["quarter", "level", "component", "value", "yoy_pct"]])

    if not frames:
        raise FileNotFoundError("No CPI component tables (3.1/3.3/3.5) found in data/raw/")

    out = pd.concat(frames, ignore_index=True).dropna(subset=["quarter"])
    out.to_csv(os.path.join(PROC, "cpi_components_yoy.csv"), index=False)
    return out


# --- 4) AFSA Insolvencies (national + state) ---------------------------------

def build_afsa():
    # National totals by quarter
    nat_path = os.path.join(RAW, "quarterly_personal_insolvencies.csv")
    nat = pd.read_csv(nat_path)

    q_col = detect_col(nat, [r"^time$", r"date", r"period", r"quarter"])
    v_col = detect_col(nat, [r"(insolv|total|cases|count|number)"])

    if q_col is None or v_col is None:
        if nat.shape[1] > 2:
            nat = nat.melt(id_vars=[nat.columns[0]], var_name="series", value_name="value")
            q_col = nat.columns[0]; v_col = "value"
        else:
            raise ValueError("Could not detect time/value in quarterly_personal_insolvencies.csv")

    afsa_nat = pd.DataFrame({
        "quarter": nat[q_col].map(to_quarter_period),
        "insolvencies_total": clean_numeric(nat[v_col]),
    }).dropna(subset=["quarter"]).sort_values("quarter")
    afsa_nat["insolvencies_yoy_pct"] = yoy(afsa_nat["insolvencies_total"])
    afsa_nat.to_csv(os.path.join(PROC, "afsa_insolvencies_quarterly.csv"), index=False)

    # State-level
    st_path = os.path.join(RAW, "regional_quarterly_time_series.csv")
    st = pd.read_csv(st_path)

    # Expect columns like: Quarter, State, Count (or similar)
    q_col = detect_col(st, [r"^time$", r"date", r"period", r"quarter"])
    s_col = detect_col(st, [r"state|region|juris"], default="State")
    v_col = detect_col(st, [r"(insolv|total|cases|count|number|value)"], default=st.columns[-1])

    st["quarter"] = st[q_col].map(to_quarter_period)
    st["state"]   = st[s_col].astype(str).str.strip()
    st["insolvencies"] = clean_numeric(st[v_col])

    afsa_state = (
        st[["quarter", "state", "insolvencies"]]
        .dropna(subset=["quarter", "state"])
        .sort_values(["state", "quarter"])
        .copy()
    )
    afsa_state["insolvencies_yoy_pct"] = afsa_state.groupby("state", group_keys=False)["insolvencies"].apply(yoy)
    afsa_state.to_csv(os.path.join(PROC, "afsa_insolvencies_state_quarterly.csv"), index=False)

    return afsa_nat, afsa_state


# --- 5) Merge macro (national) -----------------------------------------------

def build_merged_macro(unemp, cpi):
    out = pd.merge(unemp, cpi, on="quarter", how="inner").sort_values("quarter")
    out.to_csv(os.path.join(PROC, "merged_macro.csv"), index=False)
    return out


# --- 6) Final dataset (macro + insolvency, national) -------------------------

def build_final_dataset(merged_macro, afsa_nat):
    out = pd.merge(merged_macro, afsa_nat, on="quarter", how="inner").sort_values("quarter")
    # Optional: simple engineered features used later by models
    out["unemp_l1"] = out["unemployment_sa"].shift(1)
    out["cpi_yoy_l1"] = out["cpi_yoy_pct"].shift(1)
    out["insolvencies_total_l1"] = out["insolvencies_total"].shift(1)
    out.dropna(inplace=True)
    out.to_csv(os.path.join(PROC, "final_dataset.csv"), index=False)
    return out


# --- Orchestrate -------------------------------------------------------------

def main():
    print("→ Building unemployment_quarterly.csv")
    unemp = build_unemployment_quarterly()

    print("→ Building cpi_quarterly.csv")
    cpi = build_cpi_quarterly()

    print("→ Building cpi_components_yoy.csv")
    _ = build_cpi_components_yoy()

    print("→ Building AFSA insolvencies (national & state)")
    afsa_nat, afsa_state = build_afsa()

    print("→ Building merged_macro.csv")
    merged_macro = build_merged_macro(unemp, cpi)

    print("→ Building final_dataset.csv")
    _ = build_final_dataset(merged_macro, afsa_nat)

    print("✅ All processed files written to:", PROC)


if __name__ == "__main__":
    main()
```

```python
# === Part 1: Helpers (robust pickers + converters) ===
import pandas as pd
import numpy as np
import re

def to_quarter_end(dts):
    dts = pd.to_datetime(dts, errors="coerce")
    return dts.dt.to_period("Q").dt.to_timestamp("Q")

def find_first_parsable_datetime_col(df, min_frac=0.5):
    for c in df.columns:
        s = pd.to_datetime(df[c], errors="coerce")
        if s.notna().mean() >= min_frac:
            return c
    return None

def assemble_date_from_year_month(df):
    y = next((c for c in df.columns if "year" in c.lower()), None)
    m = next((c for c in df.columns if "month" in c.lower()), None)
    if y and m:
        return pd.to_datetime(df[y].astype(str)+"-"+df[m].astype(str), errors="coerce")
    return None
```

```python
# === Part 2: Unemployment ===
import os

def load_unemp_from_keystats(path):
    """Parse 'Key statistics Seasonally adjusted.csv' shaped like a matrix of 1–2 months."""
    if not os.path.exists(path):
        return pd.DataFrame(columns=["date","unemployment_sa"])

    raw = pd.read_csv(path)
    # promote row 0 to header for month columns
    promoted = raw.copy()
    new_cols = [promoted.columns[0]] + promoted.iloc[0, 1:].tolist()
    promoted = promoted.iloc[1:].reset_index(drop=True)
    promoted.columns = new_cols

    # month headers like 'Jul-25'
    month_cols = [c for c in promoted.columns[1:] if isinstance(c, str) and re.fullmatch(r"[A-Za-z]{3}-\d{2}", c)]
    if not month_cols:
        return pd.DataFrame(columns=["date","unemployment_sa"])

    # pick the unemployment row
    label_col = promoted.columns[0]
    row = promoted[label_col].astype(str).str.strip().str.lower() == "unemployment rate"
    if not row.any():
        return pd.DataFrame(columns=["date","unemployment_sa"])
    ur = promoted.loc[row].iloc[0]

    monthly = (pd.DataFrame({"month_label": month_cols,
                             "unemployment_sa": [ur[c] for c in month_cols]})
               .assign(unemployment_sa=lambda d: pd.to_numeric(d["unemployment_sa"].astype(str).str.replace("%","", regex=False),
                                                               errors="coerce"),
                       date=lambda d: pd.to_datetime(d["month_label"], format="%b-%y", errors="coerce"))
               .dropna(subset=["date","unemployment_sa"])
               .sort_values("date")
               .reset_index(drop=True))
    return monthly[["date","unemployment_sa"]]

def load_unemp_from_tidy(path):
    """Parse a tidy monthly series like 'Unemployment rate.csv'."""
    if not os.path.exists(path):
        return pd.DataFrame(columns=["date","unemployment_sa"])
    df = pd.read_csv(path)

    # date
    date_col = next((c for c in df.columns if any(k in c.lower() for k in ["month","date","period","reference month"])), None)
    if date_col:
        dt = pd.to_datetime(df[date_col], errors="coerce")
    else:
        dt = assemble_date_from_year_month(df) or (
            pd.to_datetime(df[find_first_parsable_datetime_col(df) or ""], errors="coerce")
            if find_first_parsable_datetime_col(df) else None
        )

    # value
    val_col = next((c for c in df.columns if ("unemployment" in c.lower()) or (c.lower() in ["value","rate","data value"])), None)
    if dt is None or val_col is None:
        return pd.DataFrame(columns=["date","unemployment_sa"])

    monthly = (pd.DataFrame({"date": dt, "unemployment_sa": pd.to_numeric(df[val_col], errors="coerce")})
               .dropna()
               .sort_values("date")
               .reset_index(drop=True))
    return monthly

# Load sources
ks_path   = f"{DATA_PATH}/Key statistics Seasonally adjusted.csv"
tidy_path = f"{DATA_PATH}/Unemployment rate.csv"

m_ks   = load_unemp_from_keystats(ks_path)
m_tidy = load_unemp_from_tidy(tidy_path)

# Combine: prefer tidy history; append newer months from keystats if any
if m_tidy.empty and m_ks.empty:
    raise ValueError("No unemployment data found in either file.")
if m_tidy.empty:
    monthly_u = m_ks.copy()
else:
    monthly_u = m_tidy.copy()
    if not m_ks.empty:
        cut = monthly_u["date"].max()
        newer = m_ks[m_ks["date"] > cut]
        if not newer.empty:
            monthly_u = (pd.concat([monthly_u, newer], ignore_index=True)
                         .sort_values("date").reset_index(drop=True))

# Quarterly average
monthly_u["quarter_end"] = to_quarter_end(monthly_u["date"])
unemp_q = (monthly_u.groupby("quarter_end", as_index=False)["unemployment_sa"]
                    .mean()
                    .sort_values("quarter_end"))
unemp_q["quarter"] = unemp_q["quarter_end"].dt.to_period("Q")

# Save intermediate
u_q_out = unemp_q[["quarter","unemployment_sa"]]
u_q_out.to_csv(f"{OUT_PRO}/unemployment_quarterly.csv", index=False)
print("Saved:", f"{OUT_PRO}/unemployment_quarterly.csv")
print("Monthly points:", len(monthly_u), "Quarterly points:", len(unemp_q))
```

```python
# === Part 3: CPI headline (All Groups) ===
# Uses 'Data-Table 1.csv' (ABS CPI index, quarterly)
cpi_path = f"{DATA_PATH}/Data1-Table 1.csv"
cpi_raw = pd.read_csv(cpi_path)

# find quarter & value columns
q_col = next((c for c in cpi_raw.columns if any(k in c.lower() for k in ["quarter","date","time","period","reference quarter"])), None)
v_col = next((c for c in cpi_raw.columns if ("index" in c.lower()) or (c.lower() in ["value","cpi","all groups"])), None)
assert q_col and v_col, f"Could not find quarter/value columns in {cpi_path}\nCols: {list(cpi_raw.columns)}"

cpi = cpi_raw[[q_col, v_col]].rename(columns={q_col:"quarter_raw", v_col:"cpi_index"})

def parse_quarter_col(s):
    s = s.astype(str)
    out = pd.to_datetime(s, errors="coerce")
    if out.isna().all():
        s2 = s.str.replace(r"[-\s]*Q", "Q", regex=True)
        out = pd.PeriodIndex(s2, freq="Q").to_timestamp("Q")
    return out

cpi["quarter_end"] = parse_quarter_col(cpi["quarter_raw"])
cpi = cpi.dropna(subset=["quarter_end"])
cpi["cpi_index"] = pd.to_numeric(cpi["cpi_index"], errors="coerce")
cpi = cpi.dropna(subset=["cpi_index"]).sort_values("quarter_end").drop_duplicates("quarter_end")

# YoY % change
cpi["cpi_yoy"] = cpi["cpi_index"].pct_change(4) * 100
cpi_q = cpi[["quarter_end","cpi_index","cpi_yoy"]].copy()
print(cpi_q.tail())
```

```python
# === Part 4: EDA plots with crisis shading ===
import matplotlib.pyplot as plt
from datetime import datetime

# bands for GFC & COVID (approx Australian timing)
GFC_START, GFC_END     = pd.to_datetime("2008-09-01"), pd.to_datetime("2009-06-30")
COVID_START, COVID_END = pd.to_datetime("2020-03-01"), pd.to_datetime("2021-12-31")

def shade_crises(ax):
    ax.axvspan(GFC_START, GFC_END,   alpha=0.15)
    ax.axvspan(COVID_START, COVID_END, alpha=0.15)

# Unemployment
if len(unemp_q) >= 2:
    plt.figure(figsize=(9,4))
    ax = plt.gca()
    ax.plot(unemp_q["quarter_end"], unemp_q["unemployment_sa"], linewidth=2, marker='o')
    shade_crises(ax)
    ax.set_title("Unemployment rate (SA, quarterly avg)")
    ax.set_xlabel("Quarter"); ax.set_ylabel("Percent")
    plt.tight_layout()
    path_u = f"{OUT_FIGS}/unemployment_trend_quarterly.png"
    plt.savefig(path_u, dpi=200, bbox_inches="tight"); plt.show()
else:
    # fallback to monthly
    plt.figure(figsize=(9,4))
    ax = plt.gca()
    ax.plot(monthly_u["date"], monthly_u["unemployment_sa"], linewidth=2, marker='o')
    shade_crises(ax)
    ax.set_title("Unemployment rate (SA, monthly)")
    ax.set_xlabel("Month"); ax.set_ylabel("Percent")
    plt.tight_layout()
    path_u = f"{OUT_FIGS}/unemployment_trend_monthly.png"
    plt.savefig(path_u, dpi=200, bbox_inches="tight"); plt.show()

print("Saved:", path_u)

# CPI YoY
plt.figure(figsize=(9,4))
ax = plt.gca()
ax.plot(cpi_q["quarter_end"], cpi_q["cpi_yoy"], linewidth=2, marker='o')
shade_crises(ax)
ax.set_title("CPI YoY (%)")
ax.set_xlabel("Quarter"); ax.set_ylabel("Percent")
plt.tight_layout()
path_c = f"{OUT_FIGS}/cpi_yoy_trend.png"
plt.savefig(path_c, dpi=200, bbox_inches="tight"); plt.show()
print("Saved:", path_c)
```

```python
# === Part 5: Merge macro + export ===
macro = (unemp_q.merge(cpi_q, on="quarter_end", how="inner")
                 .sort_values("quarter_end"))
macro["quarter"] = macro["quarter_end"].dt.to_period("Q")

macro_out = macro[["quarter","unemployment_sa","cpi_index","cpi_yoy"]]
macro_path = f"{OUT_PRO}/merged_macro.csv"
macro_out.to_csv(macro_path, index=False)

print("Rows:", len(macro_out))
print(macro_out.tail(8))
print("Saved:", macro_path)
```
